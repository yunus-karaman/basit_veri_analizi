```python
# ==========================================================
# üßπ OUTLIER & MISSING VALUE ANALYSIS for df_cu
# ==========================================================

# ƒ∞lgili s√ºtunlar
df_cu = df_cu[["customer_id", "sex", "customer_age", "tenure"]]

# Veri tipleri
df_cu["customer_id"] = df_cu["customer_id"].astype("object")
df_cu["sex"] = df_cu["sex"].astype("category")

# customer_age ve tenure temizleme
df_cu["customer_age"] = (
    df_cu["customer_age"].astype(str).str.extract("(\d+)").astype(float)
)
df_cu["tenure"] = pd.to_numeric(df_cu["tenure"], errors="coerce")

# 1Ô∏è‚É£ Eksik Veri Analizi
print("=== MISSING VALUE COUNTS ===")
missing_summary = df_cu.isnull().sum()
print(missing_summary, "\n")

# Eksik veri oranƒ±
missing_ratio = (df_cu.isnull().mean() * 100).round(2)
print("=== MISSING VALUE RATIOS (%) ===")
print(missing_ratio, "\n")


# 2Ô∏è‚É£ Hatalƒ± (Invalid) Veri Kontrol√º
# √ñrnek: ya≈ü veya tenure mantƒ±ksƒ±z aralƒ±kta mƒ±?
invalid_age = df_cu[(df_cu["customer_age"] < 10) | (df_cu["customer_age"] > 100)]
invalid_tenure = df_cu[(df_cu["tenure"] < 0) | (df_cu["tenure"] > 600)]

print("=== INVALID AGE RECORDS ===")
print(invalid_age)
print("\n=== INVALID TENURE RECORDS ===")
print(invalid_tenure)

# 3Ô∏è‚É£ Aykƒ±rƒ± Deƒüer (Outlier) Analizi ‚Äî IQR y√∂ntemiyle
def find_outliers_iqr(series):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    outliers = series[(series < lower) | (series > upper)]
    return outliers, lower, upper

# Age outliers
age_outliers, age_low, age_up = find_outliers_iqr(df_cu["customer_age"])
print(f"\n=== AGE OUTLIERS ({len(age_outliers)}) ===")
print(age_outliers)

# Tenure outliers
tenure_outliers, ten_low, ten_up = find_outliers_iqr(df_cu["tenure"])
print(f"\n=== TENURE OUTLIERS ({len(tenure_outliers)}) ===")
print(tenure_outliers)

# 4Ô∏è‚É£ Aykƒ±rƒ± deƒüerleri g√∂rselle≈ütir
fig, axes = plt.subplots(1, 2, figsize=(10, 4))
sns.boxplot(x=df_cu["customer_age"], ax=axes[0], color="skyblue")
axes[0].set_title("Customer Age - Outliers")
sns.boxplot(x=df_cu["tenure"], ax=axes[1], color="orange")
axes[1].set_title("Tenure - Outliers")
plt.tight_layout()
plt.show()

# 5Ô∏è‚É£ √ñzet rapor
summary = pd.DataFrame({
    "Column": ["customer_age", "tenure"],
    "Missing Values": [df_cu["customer_age"].isnull().sum(), df_cu["tenure"].isnull().sum()],
    "Invalid Values": [len(invalid_age), len(invalid_tenure)],
    "Outlier Count": [len(age_outliers), len(tenure_outliers)]
})

print("\n=== DATA QUALITY SUMMARY ===")
print(summary)
```

```python
# ==========================================================
# üßΩ CLEANING OUTLIERS & INVALID VALUES (replace with mean)
# ==========================================================

# 2Ô∏è‚É£ ƒ∞lgili s√ºtunlarƒ± se√ß ve t√ºrleri ayarla
df_cu = df_cu[["customer_id", "sex", "customer_age", "tenure"]]
df_cu["customer_id"] = df_cu["customer_id"].astype("object")
df_cu["sex"] = df_cu["sex"].astype("category")

# 3Ô∏è‚É£ customer_age ve tenure s√ºtunlarƒ±nƒ± temizle
df_cu["customer_age"] = (
    df_cu["customer_age"].astype(str).str.extract("(\d+)").astype(float)
)
df_cu["tenure"] = pd.to_numeric(df_cu["tenure"], errors="coerce")

# 4Ô∏è‚É£ INVALID (hatalƒ±) deƒüerleri bul ve NaN yap
df_cu.loc[(df_cu["customer_age"] < 10) | (df_cu["customer_age"] > 100), "customer_age"] = np.nan
df_cu.loc[(df_cu["tenure"] < 0) | (df_cu["tenure"] > 600), "tenure"] = np.nan

# 5Ô∏è‚É£ Aykƒ±rƒ± deƒüerleri (outliers) bulmak i√ßin IQR y√∂ntemi
def replace_outliers_with_mean(df, col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    mean_val = df[col].mean()
    
    # Aykƒ±rƒ±larƒ± ortalama ile deƒüi≈ütir
    df.loc[(df[col] < lower) | (df[col] > upper), col] = mean_val

# Aykƒ±rƒ± deƒüerleri d√ºzelt
replace_outliers_with_mean(df_cu, "customer_age")
replace_outliers_with_mean(df_cu, "tenure")

# 6Ô∏è‚É£ Eksik deƒüerleri ortalama ile doldur
df_cu["customer_age"].fillna(df_cu["customer_age"].mean(), inplace=True)
df_cu["tenure"].fillna(df_cu["tenure"].mean(), inplace=True)

# 7Ô∏è‚É£ Kontrol ama√ßlƒ± √∂zet
print("=== CLEANED DATA SUMMARY ===")
print(df_cu.describe(), "\n")

# 8Ô∏è‚É£ Temizlenmi≈ü veriyi kaydetmek istersen:
# df_cu.to_csv("/mnt/data/customer_details_cleaned.csv", index=False)
print("‚úÖ All invalid, missing, and outlier values replaced with mean successfully.")
```

```python
#  2. Rapor s√∂zl√ºƒü√º olu≈ütur
report = {}

# üß© 3. Eksik deƒüer analizi
missing_values = df_ba.isnull().sum()
report["missing_values"] = missing_values.to_dict()

# üß† 4. Veri t√ºr√º kontrol√º
report["dtypes"] = df_ba.dtypes.astype(str).to_dict()

# ‚è∞ 5. Hatalƒ± deƒüer kontrol√º
# Tarih formatƒ±
invalid_dates = 0
try:
    pd.to_datetime(df_ba["basket_date"], errors="raise")
except Exception:
    invalid_dates = df_ba["basket_date"].apply(lambda x: pd.to_datetime(x, errors="coerce")).isnull().sum()
report["invalid_dates"] = int(invalid_dates)

# Negatif veya sƒ±fƒ±r basket_count kontrol√º
invalid_basket_count = df_ba[df_ba["basket_count"] <= 0].shape[0]
report["invalid_basket_count"] = int(invalid_basket_count)

# ‚ö†Ô∏è 6. Aykƒ±rƒ± deƒüer analizi (IQR y√∂ntemi)
def detect_outliers_iqr(series):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = series[(series < lower_bound) | (series > upper_bound)]
    return outliers

outliers_basket_count = detect_outliers_iqr(df_ba["basket_count"])
report["outlier_basket_count_count"] = len(outliers_basket_count)
report["outlier_basket_count_values"] = outliers_basket_count.tolist()

# üßæ 7. Raporu yazdƒ±r
print("üìä Veri Kalite Raporu\n" + "-"*40)
print("Eksik Deƒüerler:")
print(missing_values)
print("\nVeri T√ºrleri:")
print(df_ba.dtypes)
print(f"\nGe√ßersiz Tarihler: {report['invalid_dates']}")
print(f"Negatif / 0 basket_count: {report['invalid_basket_count']}")
print(f"Aykƒ±rƒ± Deƒüer Sayƒ±sƒ± (basket_count): {report['outlier_basket_count_count']}")
print(f"Aykƒ±rƒ± Deƒüerler (√∂rnek): {report['outlier_basket_count_values'][:20]}")  # sadece ilk 20 tanesini g√∂ster
```

```python

# === 1Ô∏è‚É£ Tarihlere g√∂re en √ßok √ºr√ºn ===
product_by_date = (
    df_ba.groupby(["basket_date", "product_id"])["basket_count"]
    .sum()
    .reset_index()
)

# En √ßok sepete eklenen √ºr√ºnler (toplam bazda ilk 10)
top_product_summary = (
    product_by_date.groupby("product_id")["basket_count"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

plt.figure(figsize=(10,6))
top_product_summary.plot(kind='bar', color='steelblue')
plt.title("Top 10 √úr√ºn (T√ºm Tarihlerde)")
plt.xlabel("Product ID")
plt.ylabel("Toplam Sepet Adedi")
plt.grid(True)
plt.show()

# === 2Ô∏è‚É£ Tarihlere g√∂re en √ßok m√º≈üteri ===
customer_by_date = (
    df_ba.groupby(["basket_date", "customer_id"])["basket_count"]
    .sum()
    .reset_index()
)

# En √ßok alƒ±≈üveri≈ü yapan m√º≈üteriler (toplam bazda ilk 10)
top_customer_summary = (
    customer_by_date.groupby("customer_id")["basket_count"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

plt.figure(figsize=(10,6))
top_customer_summary.plot(kind='bar', color='orange')
plt.title("Top 10 M√º≈üteri (T√ºm Tarihlerde)")
plt.xlabel("Customer ID")
plt.ylabel("Toplam Sepet Adedi")
plt.grid(True)
plt.show()
```

```python


# === Tarihlere g√∂re toplam sepet sayƒ±sƒ± ===
daily_basket = (
    df_ba.groupby("basket_date")["basket_count"]
    .sum()
    .reset_index()
    .sort_values("basket_date")
)

# === Grafik ===
plt.figure(figsize=(30,8))
plt.plot(daily_basket["basket_date"], daily_basket["basket_count"], marker='o', linestyle='-', color='purple')
plt.title("Tarihlere G√∂re Toplam Sepet Sayƒ±sƒ±", fontsize=14)
plt.xlabel("Tarih", fontsize=12)
plt.ylabel("Toplam Sepet Adedi", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
